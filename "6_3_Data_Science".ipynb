```python
import string

# Функция для чтения слов из файла и их очистки
def read_words(filename):
    with open(filename, 'r') as file:
        words = file.read().split()
    cleaned_words = []
    for word in words:
        # Удаление знаков препинания и перевод в верхний регистр
        cleaned_word = word.upper().rstrip().translate(str.maketrans('', '', string.punctuation))
        cleaned_words.append(cleaned_word)
    return cleaned_words

# Функция для подсчета вхождения каждой буквы в слове (английские)
def count_letters_en(words):
    counts = {ch: 0 for ch in "ABCDEFGHIJKLMNOPQRSTUVWXYZ"}
    for word in words:
        unique_letters = set(word)  # Получаем уникальные буквы в слове
        for letter in unique_letters:
            if letter in counts:
                counts[letter] += 1
    return counts

# Функция для подсчета вхождения каждой буквы в слове (русские)
def count_letters_ru(words):
    counts = {ch: 0 for ch in "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"}
    for word in words:
        unique_letters = set(word)  # Получаем уникальные буквы в слове
        for letter in unique_letters:
            if letter in counts:
                counts[letter] += 1
    return counts

# Функция для вычисления процента слов, содержащих каждую букву
def calculate_percentages(counts, total_words):
    percentages = {}
    for letter, count in counts.items():
        percentages[letter] = (count / total_words) * 100
    return percentages

# Функция для нахождения наиболее редко используемой буквы
def find_least_frequent_letter(counts):
    smallest_count = min(counts.values())
    least_frequent_letters = [letter for letter, count in counts.items() if count == smallest_count]
    return least_frequent_letters, smallest_count

# Главная функция
def main():
    filename = 'Morning on the Wissahiccon.txt'
    words = read_words(filename)

    # Разделение на английские и русские слова
    words_en = [word for word in words if all(ch in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" for ch in word)]
    words_ru = [word for word in words if all(ch in "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ" for ch in word)]

    total_words_en = len(words_en)
    total_words_ru = len(words_ru)

    # Анализ английских слов
    if total_words_en > 0:
        counts_en = count_letters_en(words_en)
        percentages_en = calculate_percentages(counts_en, total_words_en)

        # Сортировка букв по частоте вхождения (от меньшего к большему)
        sorted_counts_en = sorted(counts_en.items(), key=lambda item: item[1])

        print("Процент слов, содержащих каждую букву (английские слова):")
        for letter, count in sorted_counts_en:
            percentage = percentages_en[letter]
            print(f'{letter}: {percentage:.2f}%')

        # Нахождение и вывод наиболее редко используемой буквы
        least_frequent_letters_en, smallest_count_en = find_least_frequent_letter(counts_en)
        print('\nНаиболее редко используемые буквы (английские слова):')
        for letter in least_frequent_letters_en:
            print(f'{letter}: {smallest_count_en} слов')

        # Выведите слова, содержащие наименее часто используемую букву
        print('\nСлова, содержащие наименее часто используемую букву (английские слова):')
        for letter in least_frequent_letters_en:
            print(f'Слова с буквой {letter}:')
            for word in words_en:
                if letter in word:
                    print(word)

    # Анализ русских слов
    if total_words_ru > 0:
        counts_ru = count_letters_ru(words_ru)
        percentages_ru = calculate_percentages(counts_ru, total_words_ru)

        # Сортировка букв по частоте вхождения (от меньшего к большему)
        sorted_counts_ru = sorted(counts_ru.items(), key=lambda item: item[1])

        print("\nПроцент слов, содержащих каждую букву (русские слова):")
        for letter, count in sorted_counts_ru:
            percentage = percentages_ru[letter]
            print(f'{letter}: {percentage:.2f}%')

        # Нахождение и вывод наиболее редко используемой буквы
        least_frequent_letters_ru, smallest_count_ru = find_least_frequent_letter(counts_ru)
        print('\nНаиболее редко используемые буквы (русские слова):')
        for letter in least_frequent_letters_ru:
            print(f'{letter}: {smallest_count_ru} слов')

        # Выведите слова, содержащие наименее часто используемую букву
        print('\nСлова, содержащие наименее часто используемую букву (русские слова):')
        for letter in least_frequent_letters_ru:
            print(f'Слова с буквой {letter}:')
            for word in words_ru:
                if letter in word:
                    print(word)

if __name__ == '__main__':
    main()
```
